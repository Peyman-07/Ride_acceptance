## Plot modifications

```python
# https://seaborn.pydata.org/tutorial/aesthetics.html

def sinplot(f=1):
    x = np.linspace(0,14,100)
    for i in range(0,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*f)
```

```python
sinplot(1)
```

```python
sns.set_style('whitegrid')
sinplot(1)
```

```python
sns.set_style('white')
sinplot(1)
```

```python
sns.set_style('ticks')
sinplot(1)
sns.despine()
```

```python
sns.set_style({'axes.facecolor':'y'})
sinplot(1)
sns.despine(left=True,bottom=True)
```

```python
sns.axes_style()
```

```python
sns.set_style({'xtick.bottom': False},{'ytick.left': False} )
sinplot(1)
sns.despine(left=True,bottom=True)
```

```python
sns.set_context('poster')
sinplot(1)
```
