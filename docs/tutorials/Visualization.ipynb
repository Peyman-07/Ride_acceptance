```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt          
import seaborn as sns
```

```python
sns.set_style("darkgrid")
```

```python
tips = sns.load_dataset('tips')
```

## relplot(relational)

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='tab10')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size', sizes=(10,200))
```

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time=np.arange(500),value=randn(500).cumsum()))
```

```python
sns.relplot(x='time', y='value', data=df, kind='line')
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='scatter')
```

```python
# if we set ci to False we will not see the confidence interval
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event', style='region', markers=True)
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri.query("event=='cue'"), kind='line')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri.query("event=='stim'"), kind='line')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', units='subject', estimator=None)
```

## catplot(categorical)

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(y='day', x='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='swarm', order=['No','Yes'])
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
g = sns.catplot(x='day', y='total_bill', data=tips, kind='violin', inner=None)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips, ax=g.ax)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True, inner='sticks')
```

```python
sns.catplot(x='sex', y='survived', data=titanic, kind='bar', hue='class')
```

```python
sns.catplot(x='deck', data=titanic, kind='count')
```

```python
sns.catplot(x='sex', y='survived', data=titanic, kind='point', hue='class')
```
