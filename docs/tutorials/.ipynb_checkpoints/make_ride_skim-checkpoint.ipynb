```python
import osmnx as ox
import networkx as nx
import random
import pandas as pd
import matplotlib.pyplot as plt          
import numpy as np
import seaborn as sns
```

```python
# Do not forget to set the radius for centeral nodes calculation
# The minimum trip distance is considered 1000m and this can be modified through utils.py: generate_demand function. 
# long attribute should be calculated according to the size of city
# Platform fare is 2$ per km

params = MaaSSim.utils.get_config('glance.json')
params.use_fixed_postions = True
params.times.patience = 600 #in second
params.dist_threshold = 10000 #in meter
params.vehicle_fixed_positions= True  #this should be fixed
params.city = "Delft, Netherlands"
params.paths.G = "../../data/graphs/Delft.graphml"
params.paths.skim = "../../data/graphs/Delft.csv"
params.paths.ride_skim = "../../data/graphs/Delft_ride.csv"
params.nP = 5
params.nV = 2
params.simTime =1
```

```python
sim= MaaSSim.simulators.simulate(params=params, logger_level = logging.WARNING)
```

```python
==============================================================================================================
```

```python
G = sim.inData.G
```

```python
central_nodes = sim.inData.stats.central_nodes
nodes, edges = ox.graph_to_gdfs(G)
index = edges.index
edges['speed'] = params.speeds.ride
#edges
```

```python
#modify the speed for cenrtal edges
for i in index:
    if i[0] in central_nodes or i[1] in central_nodes:
        edges.at[i,'speed'] = 3
```

```python
edges['travel_time'] = edges['length'] / edges['speed']
```

```python
new_G = ox.utils_graph.graph_from_gdfs(nodes, edges)
```

```python
fig, ax = ox.plot_graph(new_G, figsize=(10, 10), show=False, close=False, node_size=20, bgcolor='black', edge_color='w')
```

```python
params.city = 'Delft, Netherlands'
MaaSSim.utils.make_config_paths(params, main=None, rel = False)
```

```python
inData.G = new_G
inData.nodes = pd.DataFrame.from_dict(dict(inData.G.nodes(data=True)), orient='index')
```

```python
inData.skim_generator = nx.all_pairs_dijkstra_path_length(inData.G, weight='travel_time')
inData.skim_dict = dict(inData.skim_generator)  # filled dict is more usable
inData.skim = pd.DataFrame(inData.skim_dict).fillna(1500).T.astype(float)
```

```python
ox.save_graphml(inData.G, filepath=params.paths.G)
inData.skim.to_csv(params.paths.ride_skim, chunksize=20000000)
```

```python
def calculate_dist(_inData,source, target):
    path = nx.shortest_path(_inData.G, source, target, 
                            weight='travel_time', method='dijkstra')
    dist = 0
    for i in range(0,len(path)-1):
        d = _inData.skim.T[path[i]][path[i+1]]
        dist = dist + d
    return dist
    
```
